```python
R = 5
font1 = {'family':'serif','color':'black','size':15}
def f(x):
    return 1/x**2

x = np.linspace(R,20,3000)
y = f(x)
x1 = np.array([0,0,0,0,0])
y1 = np.array([0,0,0,0,0])
x2 = np.array([4.9,4.9,4.9,4.9,4.9,4.9])
y2 = np.array([0.001,0.005,0.020,0.025,0.035,0.040])

plt.plot(x1,y1 , linestyle = 'dashed')
plt.plot(x,y)
plt.plot(x2,y2 , linestyle = 'dashed')
plt.title("Electric Field Due to a Conducting Sphere")
plt.xlabel("R" , fontdict = font1)
plt.ylabel("E" , fontdict = font1)
plt.show()
```

```python
x = np.linspace(1,10,11)
y = np.linspace(1,10,11)
x,y = np.meshgrid(x,y)

u = x/((x**2 + y**2)**(3/2))
v = y/((x**2 + y**2)**(3/2))

plt.quiver(x,y,u,v)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Electric Field Due To a Uniformly Charged Conducting Sphere ")
plt.show
```
