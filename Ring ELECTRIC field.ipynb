```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import math
x,y,z,r,k,q = sp.symbols('x,y,z,r,k,Q', real = True, positive = True)
```

```python
f  = k*q*x/(x**2 + r**2)**sp.Rational(3/2)
f
```

```python
def f(x,r):
    return x/((x**2 + r**2)**(3/2))
```

```python
r = 2
x = np.linspace(-10,10,3000)
y = f(x,r)

plt.plot(x,f(x,r))
plt.title("E =  x/(x**2 + r**2)**sp.Rational(3/2)")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")


plt.show()
```
