```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import math
```

```python
x,y,z,σ,ε0,R = sp.symbols('x,y,z,σ,ε0,R', real = True , positive = True)
```

```python
E = (σ/(2*ε0))*(1 - x/(sp.sqrt(x**2 + R**2))).factor()
E
```

```python
sp.diff(E, x)
```

```python
def f(x, R):
    return (1/2)*(1 - (x/(np.sqrt(x**2 + R**2))))
```

```python
σ = 1
ε0 = 1
R = math.pi

x = np.linspace(-10,10,3000)
y = f(x,R)

plt.plot(x,y)
plt.title("E = (σ/(2*ε0))*(1 - x/(sp.sqrt(x**2 + R**2))")
plt.xlabel("x-Axis")
plt.ylabel("Y-Axis")


plt.show()
```

```python
σ = 1
ε0 = 1
R = math.pi

x = np.linspace(-10,10,11)
y = np.linspace(-10,10,11)

(x,y) = np.meshgrid(x,y)

u = (σ/(2*ε0))*(1 - x/(np.sqrt(x**2 + R**2)))
v = 0

plt.quiver(x,y,u,v)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Vector field due to charged disc ")
plt.show()
```
