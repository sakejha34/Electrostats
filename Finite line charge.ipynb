```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import math
```

```python
x,y,z,r,a = sp.symbols('x,y,z,r,a')
```

```python
E_perp  = (1/(x*(r+a)))*((r/sp.sqrt(x**2+r**2))+(a/sp.sqrt(x**2 + a**2)))
E_perp.factor()
```

```python
sp.diff(E_perp , x).simplify()
```

```python
E_parallel  = (1/(r+a))*((1/sp.sqrt(x**2+r**2))-(1/sp.sqrt(x**2 + a**2)))
E_parallel.factor()
```

```python
kq = 1
r = 2
a = 4
x = np.linspace(2,5,11)
y = np.linspace(2,5,11)

(x,y) = np.meshgrid(x,y)

u = (r/(x*(np.sqrt(x**2 + r**2)))) + (a/(x*(np.sqrt(x**2 + a**2)))) 
v = (1/(np.sqrt(x**2 + r**2))) - (1/(np.sqrt(x**2 + a**2)))

plt.quiver(x,y,u,v)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Electric Field Due To a Uniformly Finite Line Charge")
plt.show()
```
