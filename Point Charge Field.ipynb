```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
import math
x,y,z,k,q = sp.symbols('x y z k Q')
```

```python
x*k*q/sp.sqrt(x**2 + y**2) 
```

```python
y*k*q/sp.sqrt(x**2 + y**2)
```

```python

kq = 1
x = np.linspace(-10,10,11)
y = np.linspace(-10,10,11)

(x,y) = np.meshgrid(x,y)

u = x*(kq)/(np.sqrt(x**2 + y**2))
v = y*(kq)/(np.sqrt(x**2 + y**2))

plt.quiver(x,y,u,v)
plt.xlabel("X - Axis")
plt.ylabel("Y - Axis")
plt.title("Vector field due to a small charge Q ")
plt.show()

```
